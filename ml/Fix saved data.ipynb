```python
import pandas as pd
import numpy as np
from xgboost_general_symbol_ensemble_sharecast import *
```

```python
df = pd.read_pickle(file, compression='gzip')
```

```python
print('Filling categrical vals with phrase "NA"')
for col in ALL_CATEGORICAL_COLUMNS:
    if col in df.columns.values:
        df[col].fillna('NA', inplace=True)
```

```python
print('Optimise categorical columns')
df = optimise_df_category(df)
```

```python
print('Optimise continous columns')
df = optimise_df_continous(df)
```

```python
print('Saving processed data')
df.to_pickle(file, compression='gzip')
```

```python
temp_df = pd.DataFrame()

temp_df[['symbol', 'adjustedPrice', 'eight_week_price_return', 'eight_week_dividend_return', 'eight_week_total_return', 'eight_week_std']] = df[['symbol', 'adjustedPrice', 'eight_week_price_return', 'eight_week_dividend_return', 'eight_week_total_return', 'eight_week_std']].dropna()
less_than_zero = temp_df['eight_week_total_return'] < 0

temp_df['eight_week_volatility_perc'] = temp_df['eight_week_std'] / temp_df['adjustedPrice'] * 100
temp_df['mult_factor'] = 1 - (temp_df['eight_week_std'] / temp_df['adjustedPrice'])
temp_df.loc[less_than_zero, 'mult_factor'] = 1 + (temp_df.loc[less_than_zero, 'eight_week_std'] / temp_df.loc[less_than_zero, 'adjustedPrice'])
temp_df['ra_price_return'] = temp_df['eight_week_price_return'] * temp_df['mult_factor']
temp_df['ra_total_return'] = temp_df['ra_price_return'] + temp_df['eight_week_dividend_return']                                                  

# temp_df[temp_df['eight_week_std'] < 0.5]['eight_week_std'] = 0.5


# temp_df[temp_df['RA'].isnull()]
# lookup_df = temp_df.groupby(['symbol'])['RA'].mean()
# lookup_df
```
